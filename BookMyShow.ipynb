```python
class Theater:
    def __init__(self,rows,columns):
        self.rows = rows
        self.columns = columns
        self.booking_details = {}
        
    def show_seat(self):
        print("The available seats are!!!....( B is for booked and S is for available)")
        for row in range(self.rows + 1):
            for col in range(self.columns + 1):
                if row == 0:
                    if col == 0:
                        print(" ", end = " ")
                    else:
                        print(col, end = " ")
                else:
                    if col == 0:
                        print(row, end = " ")
                    else:
                        seat_id = str(row) + str(col)
                        if seat_id in self.booking_details:
                            print("B", end = " ")
                        else:
                            print("S", end = " ")
            print()
    
    def buy_ticket(self):
        row = int(input("Enter the row in which you want to book the seat: "))
        col = int(input("Enter the seat number in the row in which you want to book the ticket: "))
        
        if row > self.rows:
            print("Seat is not available in the theater")
            return
        
        seat_id = str(row)+str(col)
        if seat_id in self.booking_details:
            print("This seat is already booked... Please look for other seat")
            return
        
        total_seat = self.rows * self.columns
        price = 0
        
        if total_seat <= 60:
             price += 10
        else:
            half_factor = (self.rows)//2
            if row <= half_factor:
                price += 10
            else:
                price += 8
                
        should_book = int(input(f"Here is the price for seat number {col} in row {row} : {price}\n Do you want to book the ticket?\n 1. Yes\n 2.No\n Please select your choice: "))
        if should_book == 1:
            name = input("Enter your name: ")
            age = int(input("Enter your age: "))
            gender = input("M/F : ")
            phone = int(input("Enter your mobile number: "))
            details = {}
            details['Name'] = name
            details['Age'] = age
            details['Gender'] = gender
            details['Phone'] = phone
            details['Price'] = price
            self.booking_details[seat_id] = details
            print("Seat booked successfully...!")
        elif should_book ==2:
            print("You can look for other seats too... please go throgh available seats")
            
        
    def statistics(self):
        purchased_tickets = len(self.booking_details)
        
        total_seats = self.rows * self.columns
        percentage_of_booked_tickets = (purchased_tickets/total_seats)*100
    
        current_income = 0
        for seat_id in self.booking_details:
            details = self.booking_details[seat_id]
            current_income += details['Price']
        
        total_income = 0
        if total_seats <= 60:
            price = 10
            total_income += total_seats * price
        else:
            half_factor = self.rows // 2
            if self.rows <= half_factor:
                price = 10
                total_income += half_factor * price
            else:
                price= 8
                total_income += (self.rows - half_factor) * price
            
        print("Number of purchased tickets = ", purchased_tickets)
        print("Percentge of tickets booked = ", percentage_of_booked_tickets)
        print("Current income = ", current_income)
        print("Total income = ", total_income)
        
    def user_details(self):
        row = int(input("Enter the row for which you want to know the booked user info: "))
        col = int(input("Enter the seat number in the row: "))
        if row > self.rows:
            print("Seat is not available in the theater")
            return
        seat_id = str(row) + str(col)
        details = self.booking_details[seat_id]
        if seat_id in self.booking_details:
            print(f"Here is the information of the one who booked seat number {col} in row {row}")
            print(f"Name: {details['Name']}")
            print(f"Age: {details['Age']}")
            print(f"Gender: {details['Gender']}")
            print(f"Phone : {details['Phone']}")
            print(f"Booked for price: {details['Price']} $")
        else:
            print("The seat is not booked yet...!")
                  
        
rows = int(input("Enter the number of rows: "))
columns = int(input("Enter the number of number of seats in each row: "))

t = Theater(rows,columns)

while True:
    option = int(input("Choose any option from below:\n 1. Show the seats\n 2. By a ticket\n 3. Statistics\n 4. Show booked tickets user info\n 0. Exit\n Please select any option: "))
    if option == 1:
        t.show_seat()
    elif option == 2:
        t.buy_ticket()
    elif option == 3:
        t.statistics()
    elif option == 4:
        t.user_details()
    elif option == 0:
        print("Thank you... Visit Again")
        break
    elif option == "":
        print("Please selct any of the options...!")
    else:
        print("Invalid option")
    
```
